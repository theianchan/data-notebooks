## Imputation

```python
imp = Imputer(missing_values="NaN", strategy="mean", axis=0)
mtcars_imp = pd.DataFrame(imp.fit_transform(mtcars), index=mtcars.index, columns=mtcars.columns)
mtcars_imp.head(10)
```
