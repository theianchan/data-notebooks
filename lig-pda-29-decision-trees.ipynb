## Holdout validation

```python
from sklearn.cross_validation import train_test_split
```

```python
df.head(1)
```

```python
features = [
    "Pclass", 
    "Sex", 
    "Age", 
    "Cabin"
]
```

```python
df_hv = df.copy()
df_hv["Sex"] = label_encoder.fit_transform(df["Sex"])
df_hv["Cabin"] = df["Cabin"].fillna("N").str[0]
df_hv["Cabin"] = label_encoder.fit_transform(df_hv["Cabin"])
df_hv[features].head(3)
```

```python
hv_train, hv_test = train_test_split(
    df_hv,
    test_size=0.25,
    random_state=1,
    stratify=df["Survived"]
)
```

```python
hv_train_x = hv_train[["Pclass", "Sex", "Age", "Cabin"]]
hv_train_y = hv_train[["Survived"]]
hv_test_x = hv_test[["Pclass", "Sex", "Age", "Cabin"]]
hv_test_y = hv_test[["Survived"]]
```

```python
model_hv = tree.DecisionTreeClassifier()
model_hv.fit(hv_train_x, hv_train_y)
```

```python
model.score(hv_train_x, hv_train_y)
```

```python
preds = model.predict(hv_test_x)
pd.crosstab(index=hv_test_y["Survived"], columns=preds, 
            colnames=["Predicted"], rownames=["Actual"]) / len(preds)
```

## Cross validation

```python
from sklearn.cross_validation import KFold
```

```python
cv = KFold(n=len(df), n_folds=10, random_state=12)
```

```python
features = ["Sex", "Pclass", "Age", "Fare"]
```

```python
fold_accuracy = []
model = tree.DecisionTreeClassifier()
df_cv = df_hv.copy()

for train_fold, valid_fold in cv:
    train = df_cv.loc[train_fold]
    valid = df_cv.loc[valid_fold]
    model.fit(X=train[features], y=train["Survived"])
    valid_acc = model.score(X=valid[features], y=valid["Survived"])
    fold_accuracy.append(valid_acc)

print("""Accuracy per fold: {}\n
Average accuracy: {}""".format(fold_accuracy, np.mean(fold_accuracy)))
```

### Use StratifiedKFold instead of KFold

```python
from sklearn.model_selection import StratifiedKFold
```

```python
X = df_cv[features]
y = df_cv["Survived"]
```

```python
fold_accuracy = []
skf = StratifiedKFold(n_splits=10, random_state=10)

for trn_index, val_index in skf.split(X, y):
    X_trn, X_val = X.loc[trn_index], X.loc[val_index]
    y_trn, y_val = y.loc[trn_index], y.loc[val_index]
    model.fit(X=X_trn, y=y_trn)
    fold_accuracy.append(model.score(X=X_val, y=y_val))
    
print("""Fold accuracy: {}\n
Average accuracy: {}""".format(fold_accuracy, np.mean(fold_accuracy)))
```

```python
from sklearn.cross_validation import cross_val_score
```

```python
scores = cross_val_score(estimator=model,
    X=X, y=y, scoring="accuracy", cv=10
)

print("""Fold accuracy: {}\n
Average accuracy: {}""".format(scores, np.mean(scores)))
```
