```python
def add_pct_col(df, c):
    vc = pd.DataFrame(df[c].value_counts())
    vc = vc.rename(columns={c: "COUNT"})
    vc["PERCENT"] = vc["COUNT"] / len(df) * 100
    return vc

def show_categorical(df, show_head=5, show_tail=3):
    cnames = df.dtypes[df.dtypes=="object"].index
    for c in cnames:
        vc = add_pct_col(df, c)
        if (df[c].nunique() >= 10):
            vc = pd.concat([
                vc.head(show_head),
                pd.DataFrame({"COUNT": "...", "PERCENT": "..."}, index=["..."]),
                vc.tail(show_tail)
            ])
        print("{}\n{}\n".format(c, vc))
```

```python
show_categorical(df)
```
